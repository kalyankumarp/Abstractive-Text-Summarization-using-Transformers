```
import pandas as pd
from sklearn.model_selection import train_test_split
```

```
data_path = r'/content/drive/Shared drives/CS594 DNLP Project/Data'
```

```
# load from file
stories = load(open(os.path.join(data_path,'processed_cnn_data.pkl'), 'rb'))
print('Loaded Stories %d' % len(stories))
```

```
# stories[0]
```

```
temp = [list(i.values()) for i in stories]
```

```
temp2 = np.asarray(temp)
```

```
temp2[0]
```

```
temp2[:,1]
```

```
text  = [i[:2] for i in temp2[:,0]]
text[0]
```

```
summary = [i[:1] for i in temp2[:,1]]
summary[0]
```

```
df = pd.DataFrame()
df['Text'] = text
df['Summary'] = summary
df.head()
```

```
df['Text'] = df['Text'].apply(lambda x: (",".join(x)).replace(',', ' '))
df['Summary'] = df['Summary'].apply(lambda x: (",".join(x)).replace(',', ' '))
# df.dropna(inplace = True)
```

```
df.info()
```

```
df.head()
```

```
train1_df, test_df = train_test_split(df, test_size=13800, random_state=42)
train_df, val_df = train_test_split(train1_df, test_size=13800, random_state=42)
```

```
data_path = data_path+'/Data 2_2'
```

```
temp= shuffle(train1_df).iloc[:12000]
temp.to_csv(os.path.join(data_path,'2_2 Training1 data with Validation.csv'))
```

```
temp1= shuffle(train_df).iloc[:10000]
temp1.to_csv(os.path.join(data_path,'2_2 Training data.csv'))
temp2= shuffle(val_df).iloc[:1000]
temp2.to_csv(os.path.join(data_path,'2_2 Validation data.csv'))
```

```
temp3= shuffle(test_df).iloc[:1000]
temp3.to_csv(os.path.join(data_path,'2_2 Test data.csv'))
```
